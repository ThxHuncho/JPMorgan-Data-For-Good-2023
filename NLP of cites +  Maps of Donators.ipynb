```python
import pandas as pd
df = pd.read_excel('Business Account.xlsx') 
df
```

```python
df = df.dropna(subset=['City'])
df = df[df['City'] != 'Non renseigné']
df = df[df['Partner type'] == 'Partner']
df['Last Modified On'] = pd.to_datetime(df['Last Modified On'])

# Select rows with "Last Modified On" after January 1, 2021
df= df[df['Last Modified On'] > '2021-01-01']
df
```

```python
# Function to clean and standardize city names
def clean_city_name(city):
    # Check if the value is a string
    if isinstance(city, str):
        # Remove leading/trailing whitespace and convert to title case
        cleaned_city = city.strip().title()
        # Additional cleaning can be added if needed
    else:
        # If the value is not a string, leave it as is
        cleaned_city = city

    return cleaned_city
```

```python
df["City"] = df["City"].str.capitalize()
df 
```

```python
client_cities = df["City"].unique()
client_cities
```

```python
#This code is really Long to run but works!

import folium
from geopy.geocoders import Nominatim

# Initialize the map centered on Europe
europe_map = folium.Map(location=[50, 10], zoom_start=4)

# Geocode the client cities to get their coordinates
geolocator = Nominatim(user_agent="client_locator")
client_locations = []

for city in client_cities:
    location = geolocator.geocode(city)
    if location:
        client_locations.append(location)

# Add markers for each client location
for location in client_locations:
    folium.Marker(
        location=[location.latitude, location.longitude],
        popup=location.address
    ).add_to(europe_map)

# Save the map to an HTML file
europe_map.save("client_locations_map.html")
```

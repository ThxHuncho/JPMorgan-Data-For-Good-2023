```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
Businessaccount = pd.read_excel('Business Account.xlsx')
Businessaccount.head()
```

```python
Businessaccount.columns
```

```python
Businessaccount.columns = ['Business Account', 'Customer', 'Type', 'Partner type', 'Class',
       'Class Name', 'Activity sector', 'Parent Business Account',
       'Owner Name', 'Follow-up with', 'Postal Code', 'City', 'State',
       'Country ID', 'Country Name', 'Created On', 'Last Modified On']
```

```python
invoice = pd.read_excel('Invoice and Opportunities.xlsx')
```

```python
result = pd.merge(Businessaccount, invoice, on='Customer', how='inner')
```

```python
merged_df = pd.merge(invoice, Businessaccount, left_on='Customer', right_on='Business Account', how='inner')
merged_df.head(50)
```

```python
merged_df['Year'] = merged_df['Date'].dt.year
merged_df['Month'] = merged_df['Date'].dt.month
merged_df['Quarter'] = merged_df['Date'].dt.quarter
merged_df.head(50)
```

```python
Se = merged_df[['Customer_x', 'Amount', 'Year', 'Month'] ].value_counts()
Se.head()
```

```python
df_se = Se.reset_index(name='Counts')
df_se.head(10)
```

```python
data_2022 = df_se[df_se['Year'] == 2022]
data_2022.head()
```

```python
total_paid_2022 = data_2022.groupby('Customer_x')['Amount'].sum().reset_index()
total_paid_2022.columns = ['Customer_x', 'Total_Amount_2022']
total_paid_2022.head()
```

```python
total_paid_2022_sorted = total_paid_2022.sort_values(by='Total_Amount_2022', ascending=False)
total_paid_2022_sorted.head()
```
